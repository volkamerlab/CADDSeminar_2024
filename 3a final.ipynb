# Theory
## Introduction to Chemical Reaction Networks (CRNs)

### Goal
The goal of this section is to help you understand what Chemical Reaction Networks (CRNs) are, why they are significant in the field of computational biology, and how they are used to model and analyze complex biological processes

### Definition and Examples
Chemical Reaction Networks (CRNs) are frameworks used to describe and analyze the behavior of chemical reactions over time. These networks represent the interactions between molecules (like proteins, enzymes, and metabolites) using mathematical equations.

Instead of simply listing the reactions, CRNs focus on how the concentrations of different molecules change as the reactions occur. This change is often represented using Ordinary Differential Equations (ODEs), which describe the rate at which reactants are transformed into products.

### Example Reaction

$$
E + S \leftrightharpoons_{k_{-1}}^{k_1} C \xrightarrow{k_2} E + P
$$


Here, in this reaction:
- **E**: Enzyme – A molecule that speeds up a reaction without being consumed.
-**S**: Substrate – The molecule that the enzyme acts upon.
-**C**: Enzyme-Substrate Complex – A temporary state where the enzyme binds to the substrate.
-**P**: Product – The final molecule formed after the reaction

The enzyme (E) binds to the substrate (S) to form a complex (C).
The reaction happens at a rate proportional to the concentrations of E and S, governed by a rate constant (k₁).
The complex (C) can either:
Dissociate back into E and S, with a rate constant (k₋₁).
Transform into the product (P) while releasing the enzyme (E) to participate in more reactions. This step is governed by a rate constant (k₂).

## Importance
CRNs play a vital role in computational biology because they allow us to mathematically model and analyze how chemical reactions occur in living organisms. Here’s why they matter:





* **Predicting System Behavior:**
By running simulations of CRNs, scientists can predict how a biological system might respond to changes, like the introduction of a new drug or a mutation in a gene.

* **Understanding Biological Mechanisms:**
CRNs help figure out how molecules interact within a cell, such as how enzymes break down food or how cells communicate.

* **Applications in Medicine:**
They are used to study drug-receptor interactions. For example, CRNs can model how a drug binds to a receptor on a cell, helping researchers understand how effective a drug might be or what side effects it could cause.



### How Are CRNs Analyzed?
To study CRNs, scientists use a combination of mathematical techniques and computational tools. Here's a closer look at the methods:

* **Model Reaction Kinetics:**
This involves writing down equations that describe how the concentrations of molecules change over time. These equations are based on the laws of reaction rates, which depend on factors like the amount of each molecule and the speed of the reaction.

* **Simulate Dynamics:**
Scientists use computer programs to simulate these reactions. For instance, they can input initial concentrations of molecules and watch how the system evolves over time.

* **Analyze Steady-State and Transient Dynamics**: Study both the long-term behavior and immediate responses of the system.

By integrating mathematical models and computational tools, CRNs offer a systematic approach to understanding and analyzing complex biological systems.

## Basics of Dynamical Systems in CRNs

### Key Concepts

1. Chemical Species:
   - Chemical species in a CRN include reactants, intermediates, and products. These species interact through reactions, which are characterized by specific rate constants.

2. Reactions and Rate Constants:
   - Each chemical reaction in a CRN is associated with a rate constant \( k \), which quantifies how fast the reaction occurs. For example, the rate at which reactants convert to products is governed by the rate constant.

3. Reaction Networks:
   - A reaction network consists of a set of interconnected reactions. These reactions form a dynamic system where the concentration of each species changes over time.

### Dynamical Systems

### Modeling Concentrations Over Time Using ODEs

In biology and chemistry, we often want to understand how the amount of a particular molecule changes over time in a system, such as during a chemical reaction or in a cell. To do this, we use Ordinary Differential Equations (ODEs), which are mathematical tools that describe how one quantity (e.g., concentration) changes with respect to another (e.g., time). For instance, consider the dynamics of an enzyme ($[E]$), substrate ($[S]$), and the enzyme-substrate complex $[C]$ in a simple CRN:

$$
\frac{d[E]}{dt} = -k_1[E][S] + k_2[C] + k_3[C],
$$

- The term $-k_1[E][S]$ represents the consumption of enzyme and substrate to form the enzyme-substrate complex.
- The terms $k_2[C]$ and $k_3[C]$ account for the breakdown of the enzyme-substrate complex, where it either returns to the enzyme and substrate or produces the product.

### Mass-Action Kinetics
To understand where these equations come from, we use the principle of mass-action kinetics. This principle states that the rate of a reaction is proportional to the product of the concentrations of the reactants.
$$
\frac{d[C]}{dt} = k_1[E][S] - k_2[C] - k_3[C],
$$

- The term $k_1[E][S]$ represents the formation of the enzyme-substrate complex.The formation rate depends on both the concentration of the enzyme ([E]) and the substrate ([S]), as they must collide to react.

- The terms $-k_2[C]$ and $-k_3[C]$ represent the consumption of the complex, either dissociating back into the enzyme and substrate or breaking down to form the product.

These equations reflect the law of mass action, where the rate of a reaction is proportional to the concentrations of the reactants.


###Michaelis-Menten Mechanism
The Michaelis-Menten mechanism is a foundational concept in enzyme kinetics, which helps describe how enzymes work to catalyze biochemical reactions. It provides a mathematical framework for understanding how enzymes interact with substrates to produce products.
 The Two Key Steps in the Michaelis-Menten Mechanism
1. Binding Step: Formation of the Enzyme-Substrate Complex. The enzyme
(E) binds to the substrate (S) to form a temporary intermediate known as the enzyme-substrate complex (C). This step is reversible:

   $$
   E + S \rightleftharpoons C
   $$

2. Reaction: The enzyme-substrate complex \(C\) undergoes a reaction to release the enzyme \(E\) and produces the product \(P\):

   $$
   C \rightarrow E + P
   $$

This mechanism is used to describe many enzyme-catalyzed reactions and leads to a set of rate equations for the concentrations of \(E\), \(S\), and \(C\).

## Integration of ODEs


#### Euler's Method
To analyze systems like enzyme kinetics (from the Michaelis-Menten mechanism), we need to solve Ordinary Differential Equations (ODEs), which describe how concentrations of molecules like enzymes, substrates, and products change over time.

In most cases, these ODEs cannot be solved exactly using simple algebra, so we use numerical methods like Euler’s Method to approximate the solutions. Euler's method is one of the simplest numerical techniques for solving ODEs. It gives an approximate solution for an ODE of the form:
$$
\frac{dy}{dt} = f(t, y), \quad y(t_0) = y_0.
$$

**How does it work?** Euler’s method builds the solution iteratively, one step at a time, using the following formula:

$$
y_{n+1} = y_n + h \cdot f(t_n, y_n),
$$

where:
- $y_n$: The approximate value of y at step n (current time point $t_n$)
- $h$: Step size, which determines how far forward in time you move with each step.
- $f(t_n, y_n)$: Slope (right-hand side of the ODE) evaluated at $t_n$, based on ODE.
- $y_n$ +1: The value of y at the next time step $t_n$ +1


### Step by Step explanation:
1. It starts at the initial valuey(t0) = y0
2.  compute the slope $f(t_n, y_n)$ at the current point using the ODE.
3. Move forward in time by adding the product of the slope and the step size (*h*. *f*($t_n$, $y_n$)) to the current value.
4. Repeat for the desired number of steps or until the desired time is reached.


### Shortcomings of Euler's Method
While Euler’s method is simple and easy to use, it has some limitations, especially for complex or stiff systems like enzyme reactions. Let’s explore these shortcomings in more detail:

1. Assumption of Linearity:
   - Euler’s method assumes that the rate of change (slope) remains constant within each step. This is only true for linear systems; for highly nonlinear systems (e.g., enzyme kinetics with feedback loops), this assumption can lead to significant inaccuracies.

2. Step Size Sensitivity:
   
  **The step size (h) greatly affects the accuracy:**
* Large step sizes: The approximation becomes less accurate because it overshoots the actual curve.

* Small step sizes: Improves accuracy but increases computational cost, as more steps are needed to cover the same time span.


3. Accumulation of Errors:
   - At each step, Euler’s method introduces a small error, called the local truncation error, which can add up over time leading to a global error deviating from the real solution.

4. Inefficiency for Stiff Systems:
   - For stiff ODEs, Euler's method often requires very small step sizes to maintain stability and accuracy making it computationally inefficient.

## Exponential Decay and Euler’s Method
**What Is Exponential Decay?**

Exponential decay describes a process where a quantity decreases at a rate proportional to its current value. In simpler terms:
"The faster something decreases, the smaller it gets over time."

This type of behavior is common in nature and science, such as:

* Radioactive decay: The amount of a radioactive substance decreases over time.
* Population dynamics: Certain populations shrink due to unfavorable conditions.
* Drug metabolism: The concentration of a drug in the bloodstream decreases as the body processes it.



### How It Works

The governing equation is:

$$
\frac{dy}{dt} = k \cdot y.
$$

Here,
* dy/dt : Rate of change of y over time
* K: a positive constant, called decay rate, that determines how fast the quantity decreases.
* y: the value of the quantity at a given time t

**If for example, k = 0.5, the equation becomes,**

$$
\frac{dy}{dt} = -0.5 \cdot y.
$$

This means that at any time, the quantity \( y(t) \) shrinks at half the value of its current size.



### How do We Solve It

Using Euler’s Method, we calculate $y(t)$ step by step:

1. Start with an initial value: $y_0$.

2. Use the formula:

   $$
   y_{\text{next}} = y_{\text{current}} + h \cdot \text{rate of change},
   $$

   where:
   - $y_n$: Approximate value of $y$ at time $t_n$,
   - $h$: Step size,
   - $f(t_n, y_n)$: Slope (right-hand side of the ODE) evaluated at $t_n$.

3. Repeat to get values of $y$ at later times.

### Step-by-Step Process

1. Begin at $t_0$ with $y_0$.
2. Compute the rate of change at $y_{\text{current}}$ using $-0.5 \cdot y_{\text{current}}$.
3. Update $y_{\text{next}}$ using the formula:

   $$
   y_{\text{next}} = y_{\text{current}} + h \cdot (-0.5 \cdot y_{\text{current}}).
   $$

4. Increment $t$ by the step size $h$ and repeat.

This iterative process allows us to approximate the solution $y(t)$ over time.


## Enzymatic Reaction Dynamics

Four key components of an enzymatic reaction change over time are:



### Components

1. Enzyme (E):
   - The enzyme facilitates the reaction by binding to the substrate, forming a temporary complex.
   - Its concentration decreases slightly at first but recovers as it is recycled.

2. Substrate (S):
   - The substrate is the starting material that reacts with the enzyme.
   - Its concentration decreases steadily as it is converted into the product.

3. Enzyme-Substrate Complex (C):
   - This complex forms when the enzyme binds to the substrate.
   - Its concentration increases initially, then stabilizes as the reaction reaches a balance.

4. Product (P):
   - The product is the outcome of the reaction.
   - Its concentration increases steadily as the substrate is consumed.



### What’s Happening in the Reaction?

#### 1. Binding

The enzyme (E) and substrate (S) combine to form the enzyme-substrate complex (\(C):

$$
E + S \rightarrow C
$$



#### 2. Reaction

The enzyme-substrate complex (C) breaks down to release the enzyme (E) and produce the product (P):

$$
C \rightarrow E + P
$$



#### 3. Recycling

The enzyme (E) is free to bind to more substrate, allowing the process to repeat.



This step-by-step process highlights how the enzymatic reaction progresses over time, with each component dynamically changing as the reaction proceeds.

### Interactive Simulation: Enzyme-Substrate Reaction

This Python script simulates the dynamics of an enzyme-substrate reaction using a system of ordinary differential equations (ODEs). The key features are:

1. **Chemical Reaction Network (CRN):**  
   The reaction follows these steps:  

$$
E + S \rightleftharpoons_{k_1}^{k_2} C \xrightarrow{k_3} E + P
$$
   where:  
   - $E$: Enzyme  
   - $S$: Substrate  
   - $C$: Enzyme-Substrate Complex  
   - $P$: Product  
   - $k_1, k_2, k_3$: Rate constants for binding, unbinding, and product formation.

2. **ODE Model:**  
   The system of ODEs is defined as:  
   $\frac{dE}{dt} = -k_1 E S + k_2 C + k_3 C$  
   $\frac{dS}{dt} = -k_1 E S + k_2 C$  
   $\frac{dC}{dt} = k_1 E S - (k_2 + k_3) C$  
   $\frac{dP}{dt} = k_3 C$

3. **Simulation Parameters:**  
   - Initial concentrations: $[E_0] = 1.0$, $[S_0] = 10.0$, $[C_0] = 0.0$, $[P_0] = 0.0$  
   - Time points: Simulated over $t = 0$ to $t = 50$ with 500 intervals.

4. **Interactive Exploration:**  
   - Users can adjust rate constants ($k_1$, $k_2$, $k_3$) using sliders.  
   - The resulting concentration dynamics of $E$, $S$, $C$, and $P$ are visualized over time.



```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

# Define the CRN model
def crn_model(y, t, k1, k2, k3):
    E, S, C, P = y
    dE_dt = -k1 * E * S + k2 * C + k3 * C
    dS_dt = -k1 * E * S + k2 * C
    dC_dt = k1 * E * S - (k2 + k3) * C
    dP_dt = k3 * C
    return [dE_dt, dS_dt, dC_dt, dP_dt]

# Initial conditions
E0, S0, C0, P0 = 1.0, 10.0, 0.0, 0.0

# Time points
t = np.linspace(0, 50, 500)

# Interactive function
def simulate_crn(k1=0.1, k2=0.05, k3=0.1):
    solution = odeint(crn_model, [E0, S0, C0, P0], t, args=(k1, k2, k3))
    E, S, C, P = solution.T

    plt.figure(figsize=(8, 5))
    plt.plot(t, E, label="[E] (Enzyme)")
    plt.plot(t, S, label="[S] (Substrate)")
    plt.plot(t, C, label="[C] (Complex)")
    plt.plot(t, P, label="[P] (Product)")
    plt.xlabel("Time")
    plt.ylabel("Concentration")
    plt.title("Enzyme-Substrate Reaction Dynamics")
    plt.legend()
    plt.show()

# Create interactive sliders
interact(simulate_crn, k1=(0.01, 1.0, 0.01), k2=(0.01, 1.0, 0.01), k3=(0.01, 1.0, 0.01));
```

### **Implementation of Eulers method using Exponential Decay as a case study**

This code demonstrates its implementation using exponential decay as a case study. The goal is to solve the differential equation
$d\frac{dy}{dt} = -\lambda y$, where $\lambda$ is the decay rate, and compare the numerical results with the exact solution
$y(t) = y_0 e^{-\lambda t}$.

The code includes:

1. **Euler's Method Implementation**: A function to compute approximate solutions using a specified step size.
2. **Decay Function**: Defines the decay rate for the problem.
3. **Plotting and Error Analysis**: Visualization of numerical solutions, exact solutions, and their errors.
4. **Interactive Widget**: Allows users to explore the impact of parameters like step size, decay rate, and simulation time interactively.



```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

# Define the Euler method function
def euler_method(f, y0, t_range, h):
    """
    Implements the Euler method for solving ODEs.

    Args:
        f: Function that returns dy/dt
        y0: Initial value
        t_range: Tuple of (start_time, end_time)
        h: Step size

    Returns:
        t: Array of time points
        y: Array of solution values
    """
    t = np.arange(t_range[0], t_range[1], h)
    y = np.zeros(len(t))
    y[0] = y0

    for i in range(1, len(t)):
        y[i] = y[i-1] + h * f(y[i-1])

    return t, y

# Define the decay function
def decay_function(decay_rate):
    """Returns a function that computes dy/dt = -λy."""
    return lambda y: -decay_rate * y

# Function to plot solutions
def plot_solutions(h, decay_rate, t_max):
    """
    Plots both Euler method and exact solutions.

    Args:
        h: Step size
        decay_rate: Decay rate (lambda)
        t_max: Maximum time
    """
    # Initial conditions
    y0 = 1.0
    t_range = (0, t_max)

    # Get numerical solution
    t_euler, y_euler = euler_method(decay_function(decay_rate), y0, t_range, h)

    # Compute exact solution
    t_exact = np.linspace(t_range[0], t_range[1], 200)
    y_exact = y0 * np.exp(-decay_rate * t_exact)

    # Create figure with two subplots
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))

    # Plot solutions
    ax1.plot(t_euler, y_euler, 'b.-', label='Euler Method', markersize=8)
    ax1.plot(t_exact, y_exact, 'r--', label='Exact Solution')
    ax1.grid(True)
    ax1.set_xlabel('Time (t)')
    ax1.set_ylabel('y(t)')
    ax1.set_title(f'Exponential Decay: dy/dt = -{decay_rate}y\nStep size h = {h}')
    ax1.legend()

    # Plot absolute error
    error = np.abs(y_euler - np.exp(-decay_rate * t_euler))
    ax2.plot(t_euler, error, 'g.-', label='Absolute Error')
    ax2.grid(True)
    ax2.set_xlabel('Time (t)')
    ax2.set_ylabel('|Error|')
    ax2.set_title('Absolute Error vs Time')
    ax2.legend()

    plt.tight_layout()
    plt.show()

    # Print some statistics
    print(f"\nMaximum absolute error: {error.max():.6f}")
    print(f"Average absolute error: {error.mean():.6f}")

# Create interactive widget
def interactive_euler():
    """Creates an interactive plot with widgets for parameters."""
    interact(plot_solutions,
            h=widgets.FloatSlider(min=0.01, max=0.5, step=0.01, value=0.1,
                                description='Step Size (h):'),
            decay_rate=widgets.FloatSlider(min=0.1, max=2.0, step=0.1, value=0.5,
                                         description='Decay Rate (λ):'),
            t_max=widgets.FloatSlider(min=1.0, max=20.0, step=0.5, value=10.0,
                                    description='Max Time:'));

# Example usage with specific values
def print_example_steps():
    """Prints example steps for a specific case."""
    h = 0.1
    decay_rate = 0.5
    t_range = (0, 1)
    y0 = 1.0

    t, y = euler_method(decay_function(decay_rate), y0, t_range, h)

    print("\nExample steps (first 5):")
    print("  t      y(numerical)    y(exact)        error")
    print("-" * 50)

    for i in range(5):
        y_exact = y0 * np.exp(-decay_rate * t[i])
        error = abs(y[i] - y_exact)
        print(f"{t[i]:5.2f}   {y[i]:12.6f}   {y_exact:12.6f}   {error:10.6f}")

# Run the interactive simulation
print("Interactive Euler Method Demonstration")
print("====================================")
interactive_euler()

# Print example steps
print_example_steps()
```

### **Application : Calcium Oscillations**

This Python code simulates calcium oscillations using a system of differential equations. The model describes the dynamics of calcium concentration $[\text{Ca}^{2+}]$ in response to various parameters:

- $k_{\text{in}}$: Rate of calcium inflow.
- $k_{\text{out}}$: Rate of calcium outflow.
- $k_{\text{feedback}}$: Feedback strength of calcium.
- $\text{threshold}$: Threshold for feedback activation.

The system of equations is solved using `odeint` from `scipy.integrate`. Key features include:

1. **Visualization**: Plots of calcium concentration over time and the phase space diagram ($[\text{Ca}^{2+}]$ vs. $\frac{d[\text{Ca}^{2+}]}{dt}$).
2. **Interactive Exploration**: Widgets allow users to adjust $k_{\text{in}}$, $k_{\text{out}}$, $k_{\text{feedback}}$, and $\text{threshold}$ to explore their impact.
3. **Oscillation Analysis**: Identifies peaks to compute average period, amplitude, and peak-to-peak variation.

Below is the implementation:



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from ipywidgets import widgets
from IPython.display import display, clear_output

def calcium_system(y, t, k_in, k_out, k_feedback, threshold):
    """
    System of differential equations for calcium oscillations
    """
    Ca, dCa = y
    d2Ca = k_in - k_out * Ca + k_feedback * (Ca**2 / (threshold**2 + Ca**2)) - 0.1 * dCa
    return [dCa, d2Ca]

# Create individual widgets
k_in_slider = widgets.FloatSlider(
    value=2.0,
    min=0.5,
    max=4.0,
    step=0.1,
    description='k_in:',
    continuous_update=False
)

k_out_slider = widgets.FloatSlider(
    value=1.0,
    min=0.1,
    max=2.0,
    step=0.1,
    description='k_out:',
    continuous_update=False
)

k_feedback_slider = widgets.FloatSlider(
    value=3.0,
    min=1.0,
    max=5.0,
    step=0.1,
    description='k_feedback:',
    continuous_update=False
)

threshold_slider = widgets.FloatSlider(
    value=1.5,
    min=0.5,
    max=3.0,
    step=0.1,
    description='threshold:',
    continuous_update=False
)

def plot_system(k_in, k_out, k_feedback, threshold):
    # Clear previous output
    clear_output(wait=True)

    # Display widgets again to keep them visible
    display(k_in_slider, k_out_slider, k_feedback_slider, threshold_slider)

    # Initial conditions and time points
    y0 = [0.5, 0.0]
    t = np.linspace(0, 50, 1000)

    # Calculate solution
    solution = odeint(calcium_system, y0, t, args=(k_in, k_out, k_feedback, threshold))

    # Create plots
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

    # Plot concentration
    ax1.plot(t, solution[:, 0], 'b-', linewidth=2)
    ax1.set_xlabel('Time (s)')
    ax1.set_ylabel('[Ca²⁺] (μM)')
    ax1.set_title('Calcium Oscillations')
    ax1.grid(True, alpha=0.3)

    # Plot phase space
    ax2.plot(solution[:, 0], solution[:, 1], 'r-', linewidth=1)
    ax2.set_xlabel('[Ca²⁺] (μM)')
    ax2.set_ylabel('d[Ca²⁺]/dt')
    ax2.set_title('Phase Space Diagram')
    ax2.grid(True, alpha=0.3)

    # Add direction arrow
    arrow_idx = 200
    ax2.arrow(solution[arrow_idx, 0], solution[arrow_idx, 1],
              solution[arrow_idx+1, 0] - solution[arrow_idx, 0],
              solution[arrow_idx+1, 1] - solution[arrow_idx, 1],
              head_width=0.05, head_length=0.1, fc='k', ec='k')

    plt.tight_layout()
    plt.show()

    # Print analysis
    analyze_oscillations(t, solution)

def analyze_oscillations(t, solution):
    peaks = []
    for i in range(1, len(solution)-1):
        if solution[i-1, 0] < solution[i, 0] > solution[i+1, 0]:
            peaks.append((t[i], solution[i, 0]))

    if len(peaks) >= 2:
        periods = [peaks[i+1][0] - peaks[i][0] for i in range(len(peaks)-1)]
        amplitudes = [peak[1] for peak in peaks]

        print("\nOscillation Analysis:")
        print(f"Number of peaks: {len(peaks)}")
        print(f"Average period: {np.mean(periods):.2f} seconds")
        print(f"Average amplitude: {np.mean(amplitudes):.2f} μM")
        print(f"Peak-to-peak variation: {np.std(amplitudes):.3f} μM")

# Connect the widgets to the plot function
def on_value_change(change):
    plot_system(
        k_in_slider.value,
        k_out_slider.value,
        k_feedback_slider.value,
        threshold_slider.value
    )

k_in_slider.observe(on_value_change, names='value')
k_out_slider.observe(on_value_change, names='value')
k_feedback_slider.observe(on_value_change, names='value')
threshold_slider.observe(on_value_change, names='value')

# Initial display
display(k_in_slider, k_out_slider, k_feedback_slider, threshold_slider)
plot_system(2.0, 1.0, 3.0, 1.5)
```

For instance, the enzyme-substrate reaction can be expressed as $E + S \rightleftharpoons C$ (binding) and $C \rightarrow E + P$ (product formation). This mathematical framework allows precise modeling of reaction kinetics and system behaviors.

CRNs are widely applied in biological systems like glycolysis pathways, signal transduction, and regulatory feedback loops.
